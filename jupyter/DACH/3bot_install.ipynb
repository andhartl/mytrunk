## 3bot local install (kickstart) document

This document is intended to work as a starting script and has reference pointers to the GitHub repositories that are the single version of the truth

### Step 1 Install 3bot docker image

Recommended is to do the container installation.  First download the docker container image found here:
```
https://hub.docker.com/r/threefoldtech/3bot
```

Pull the image:

### Step 2 Make sure you haev an identity

Make sure your acccount has an ssh identity:
```
ssh-add -L
```

If not load it:
```
eval `ssh-agent`
ssh-add $HOME/.ssh/id_rsa
ssh-add -L
```

### Step 3 Get the installation script and execute it

To get the install/configuration script and make it executable and rn the install / configure script:
```
curl https://raw.githubusercontent.com/threefoldtech/jumpscaleX_core/development/install/jsx.py?$RANDOM > /tmp/jsx;
chmod +x /tmp/jsx;
```

Make sure your acccount has an ssh identity:
```
ssh-add -L
<<long hash>>
```

Then configure / install the container:
```
/tmp/jsx container-install
```

### Step 4 Log into you 3bot

Then you can get into the container:
```
root@happy:~# /tmp/jsx container-shell

```
or connect to it via ssh
```
root@happy:~# ssh root@localhost -p 9000 -A
OK
3BOTDEVEL:3bot:~: 
```

If the above failed, here's the installation documentation:
```
https://github.com/threefoldtech/jumpscaleX_core/blob/development/docs/Installation/README.md
```

when success full completed
```
kosmos
```
You will then start an interactive python shell with all the jumpscale libraries loaded
```
JSX>
```
